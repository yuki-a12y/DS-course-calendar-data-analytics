# create_sample_dataでやることの例
①カレンダーの10AMといった画像を読み込み

②(256, 64)にリサイズ→2値化処理

③左端から(64, 64)で切り出す

④x=32に線を引く

⑤線が文字列を分割しているか、自分で判断する

⑥文字列を分割できていたら1,できていなかったら0を書き込む（ラベリングしている）

⑦⑥に従って、それぞれのフォルダに保存する

⑧切り出す部分を1pxずつ左方向にずらしながら、③～⑦を繰り返す。

リサイズする大きさ、2値化処理のしきい値、ラベル、切り出す部分を移動する量は指定できる

切り出しは、とりあえず正方形で行う

生成された画像は、output_path/(ラベル)/(ラベル)-(4桁の番号).jpgで保存される。

4桁の番号は0001から順に生成される。既に、ディレクトリ内にファイルがある場合は、ファイル数＋１の番号から順に生成される。

```python
create_sample_data = create_sample_data.CreateSampleData(label_list=['0', '1'], thresh=128, x_step=1, convert_size=(256,64))
```



> label_list:[str, str,.........]

ラベルを指定できる。



> thresh:int

2値化処理のしきい値。

指定した値より大きい色値が255、それ以下が0になります。



> x_step:int

画像を切り取るバッチを1度に移動させるpxです。



> convertsize:tuple(int,int)

リサイズする大きさです。
