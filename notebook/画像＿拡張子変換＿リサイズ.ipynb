## pngからjpgの変換のクオリティー、リサイズの大きさを任意の値にする


```python
convert = img_convert.ImgConvert(to_jpg_quality=1, convert_size=(1000,1000)) #質１、大きさ(1000,1000)でインスタンスの作成
```

```python
input_path = '/content/sample_data/png-original' #オリジナルのデータのディレクトリのパス
output_path = '/content/sample_data/jpg-quality1-bigsize' #保存先のディレクトリのパス、作っていないディレクトリでも勝手に作成してくれる
```

```python
convert.png_to_jpg_resize(input_path, output_path)
```

```python
img_converted = Image.open('/content/sample_data/jpg-quality1-bigsize/apple1.jpg')
plt.imshow(img_converted)
```
